```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("apple_stock.csv")
df
```

```python
df['Date'].apply(pd.to_datetime)
```

```python
df['Date'] = df['Date'].apply(pd.to_datetime)
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df['month'] = df.index.month
```

```python
df.head()
```

```python
df.groupby('month').agg(sum)
```

```python
df.groupby(df.index.year).sum()
```

```python
df.groupby([df.index.year, df.index.month]).agg(sum)
```

```python
# time series offset aliases
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='A').mean()['2009']
```

```python
def first_day(sample):
    return sample[0]
```

```python
df.resample(rule='A').apply(first_day)
```

```python
df['Close'].resample('A').mean().plot(kind='bar')
```

```python
df['Open']['2015-1':'2015-12'].resample('M').max().plot(kind='bar', figsize=(15,8))
```

```python
daily_daterange = pd.date_range(start=datetime(2019,9,1), end=datetime(2019,10,31), freq='B')
```

```python
daily_dataset = pd.DataFrame(
    data={'value':np.random.rand(len(daily_daterange))},
    index=daily_daterange)
```

```python
daily_dataset.resample('W-MON').min()
```

```python
daily_dataset.resample('M').min()
```
